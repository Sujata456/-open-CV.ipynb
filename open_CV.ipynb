```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from google.colab.patches import cv2_imshow

from google.colab import files
uploaded=files.upload()
```

```python
# step 3 Read the uploaded image
filename=list(uploaded.keys())[0]
image=cv2.imread(filename)
cv2_imshow(image)
```

```python
# step 4
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
cv2_imshow(gray)
```

```python
# step 5:resize image
# 1 picture
resized=cv2.resize(image,(300,300))
cv2_imshow(resized)

# multiple picture
a=(image,gray)
for i in a:
  resized=cv2.resize(i, (300,300))
  cv2_imshow(resized)
```

```python
# step 6
# (24,255,190) rgb value  (250,250)(box value)


annotated=image.copy()
cv2.rectangle(annotated, (50,30), (250,100), (24,255,190), 3)
cv2.putText(annotated, 'my box', (40,45), cv2.FONT_HERSHEY_COMPLEX, 1, (255,10,89), 2)
cv2_imshow(annotated)
```

```python
# step 7

cv2.imwrite('output_image.jpg', annotated)
print("Image saved as output_image.jpg")
```

```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator, load_img, img_to_array
```

```python
# step 3
uploaded=files.upload()
filename=list(uploaded.keys())[0]
```

```python
# step 4
image=load_img(filename, target_size=(224,224)) # resize image
img_array=img_to_array(image)
img_array=np.expand_dims(img_array, axis=0)  # Convert to batch of i
```

```python
# define the augmentation generator
datagen=ImageDataGenerator(
    rotation_range=17,
    width_shift_range=1.9,
    height_shift_range=0.2,
    shear_range=0.2,
    zoom_range=0.2,
    horizontal_flip=True,
    brightness_range=[0.5, 1.5]
)

# Generate augmanted image
i=0
for batch in datagen.flow(img_array, batch_size=1):
  plt.figure(figsize=(2,2))
  plt.imshow(batch[0].astype('uint8'))
  plt.axis('off')
  i+=1
  if i>5:
    break
```
